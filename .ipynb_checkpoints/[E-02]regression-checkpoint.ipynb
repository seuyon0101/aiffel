## (2) Î™®Îç∏Ïóê ÏûÖÎ†•Ìï† Îç∞Ïù¥ÌÑ∞ X Ï§ÄÎπÑÌïòÍ∏∞

- df_XÏóê ÏûàÎäî Í∞íÎì§ÏùÑ numpy arrayÎ°ú Î≥ÄÌôòÌï¥ÏÑú Ï†ÄÏû•Ìï¥Ï£ºÏÑ∏Ïöî.

## (3) Î™®Îç∏Ïóê ÏòàÏ∏°Ìï† Îç∞Ïù¥ÌÑ∞ y Ï§ÄÎπÑÌïòÍ∏∞

- df_yÏóê ÏûàÎäî Í∞íÎì§ÏùÑ numpy arrayÎ°ú Î≥ÄÌôòÌï¥ÏÑú Ï†ÄÏû•Ìï¥Ï£ºÏÑ∏Ïöî.

```python
df_X = np.array(df_X)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(df_X)
df_X = scaler.transform(df_X)
```

```python
df_X
```

## (4) train Îç∞Ïù¥ÌÑ∞ÏôÄ test Îç∞Ïù¥ÌÑ∞Î°ú Î∂ÑÎ¶¨ÌïòÍ∏∞

- XÏôÄ y Îç∞Ïù¥ÌÑ∞Î•º Í∞ÅÍ∞Å train Îç∞Ïù¥ÌÑ∞ÏôÄ test Îç∞Ïù¥ÌÑ∞Î°ú Î∂ÑÎ¶¨Ìï¥Ï£ºÏÑ∏Ïöî.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_X, df_y, test_size = 0.3, random_state =23)

print(X_train.shape, X_test.shape)
print(y_train.shape,y_test.shape)
```

## (5) Î™®Îç∏ Ï§ÄÎπÑÌïòÍ∏∞

- ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞ Í∞úÏàòÏóê ÎßûÎäî Í∞ÄÏ§ëÏπò WÏôÄ bÎ•º Ï§ÄÎπÑÌï¥Ï£ºÏÑ∏Ïöî.
- Î™®Îç∏ Ìï®ÏàòÎ•º Íµ¨ÌòÑÌï¥Ï£ºÏÑ∏Ïöî.

```python
np.random.seed(0)
W = np.random.rand(10)
b = np.random.rand()
print(W)
print(b)
```

```python
def regression_model(X,W,b) :
    y = np.dot(X,W) + b
    return y
```

```python
y_hat = regression_model(X_train, W,b)
```

## (6) ÏÜêÏã§Ìï®Ïàò loss Ï†ïÏùòÌïòÍ∏∞

- ÏÜêÏã§Ìï®ÏàòÎ•º MSE Ìï®ÏàòÎ°ú Ï†ïÏùòÌï¥Ï£ºÏÑ∏Ïöî.

```python
def MSE(a,b):
    MSE = ((a - b) ** 2).mean()
    return MSE
```

```python
def RMSE(mse):
    return mse**0.5
```

```python
MSE(y_hat,y_train)
```

## (7) Í∏∞Ïö∏Í∏∞Î•º Íµ¨ÌïòÎäî gradient Ìï®Ïàò Íµ¨ÌòÑÌïòÍ∏∞

- Í∏∞Ïö∏Í∏∞Î•º Í≥ÑÏÇ∞ÌïòÎäî gradient Ìï®ÏàòÎ•º Íµ¨ÌòÑÌï¥Ï£ºÏÑ∏Ïöî.

```python
def gradient_descent(X,predicted, y) :
    N = len(X)
    dW = 1/N * 2 * X.T.dot(predicted-y)
    db = 2 * (predicted - y).mean()
    return dW, db
```

## (8) ÌïòÏù¥Ìçº ÌååÎùºÎØ∏ÌÑ∞Ïù∏ ÌïôÏäµÎ•† ÏÑ§Ï†ïÌïòÍ∏∞

- ÌïôÏäµÎ•†, learning rate Î•º ÏÑ§Ï†ïÌï¥Ï£ºÏÑ∏Ïöî
- ÎßåÏïΩ ÌïôÏäµÏù¥ Ïûò ÎêòÏßÄ ÏïäÎäîÎã§Î©¥ learning rate Í∞íÏùÑ ÌïúÎ≤à Ïó¨Îü¨ Í∞ÄÏßÄÎ°ú ÏÑ§Ï†ïÌïòÎ©∞ Ïã§ÌóòÌï¥ Î≥¥ÏÑ∏Ïöî.

```python
LearningRate = 0.01
```

## (9) Î™®Îç∏ ÌïôÏäµÌïòÍ∏∞

- Ï†ïÏùòÎêú ÏÜêÏã§Ìï®ÏàòÏôÄ Í∏∞Ïö∏Í∏∞ Ìï®ÏàòÎ°ú Î™®Îç∏ÏùÑ ÌïôÏäµÌï¥Ï£ºÏÑ∏Ïöî.
- lossÍ∞íÏù¥ Ï∂©Î∂ÑÌûà Îñ®Ïñ¥Ïßà ÎïåÍπåÏßÄ ÌïôÏäµÏùÑ ÏßÑÌñâÌï¥Ï£ºÏÑ∏Ïöî.
- ÏûÖÎ†•ÌïòÎäî Îç∞Ïù¥ÌÑ∞Ïù∏ XÏóê Îì§Ïñ¥Í∞ÄÎäî ÌäπÏÑ± Ïª¨ÎüºÎì§ÏùÑ Î™á Í∞ú ÎπºÎèÑ Í¥úÏ∞ÆÏäµÎãàÎã§. Îã§ÏñëÌïú Îç∞Ïù¥ÌÑ∞Î°ú Ïã§ÌóòÌï¥ Î≥¥ÏÑ∏Ïöî.

```python
epoch = 2000
losses = []

for i in range(epoch+1) :
    prediction = regression_model(X_train, W, b)
    loss = MSE(y_train ,prediction)
    dW, db = gradient_descent(X_train, prediction, y_train)
    W -= LearningRate * dW
    b -= LearningRate * db
    losses.append(loss)
    
    if i % 100 == 0 :
        print(f'for epoch {i}, loss : {loss}')
```

## (10) test Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌïú ÏÑ±Îä• ÌôïÏù∏ÌïòÍ∏∞

- test Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌïú ÏÑ±Îä•ÏùÑ ÌôïÏù∏Ìï¥Ï£ºÏÑ∏Ïöî.

```python
test_prediction = regression_model(X_test,W,b)
final_loss = MSE(test_prediction, y_test)
final_loss
```

```python
RMSE(final_loss)
```

## (11) Ï†ïÎãµ Îç∞Ïù¥ÌÑ∞ÏôÄ ÏòàÏ∏°Ìïú Îç∞Ïù¥ÌÑ∞ ÏãúÍ∞ÅÌôîÌïòÍ∏∞

- xÏ∂ïÏóêÎäî X Îç∞Ïù¥ÌÑ∞Ïùò Ï≤´ Î≤àÏß∏ Ïª¨ÎüºÏùÑ, yÏ∂ïÏóêÎäî Ï†ïÎãµÏù∏ target Îç∞Ïù¥ÌÑ∞Î•º ÎÑ£Ïñ¥ÏÑú Î™®Îç∏Ïù¥ ÏòàÏ∏°Ìïú Îç∞Ïù¥ÌÑ∞Î•º ÏãúÍ∞ÅÌôîÌï¥ Ï£ºÏÑ∏Ïöî.

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
plt.scatter(df_X[:,3], df_y)
plt.scatter(df_X[:,3], regression_model(df_X,W,b))

plt.show()
```

## (5) LinearRegression Î™®Îç∏ ÌïôÏäµ

```python
from sklearn.linear_model import LinearRegression
```

```python
linearmodel = LinearRegression()
linearmodel.fit(X_train, y_train)
```

## (6) ÌïôÏäµÎêú Î™®Îç∏Î°ú X_testÏóê ÎåÄÌïú ÏòàÏ∏°Í∞í Ï∂úÎ†• Î∞è ÏÜêÏã§Ìï®ÏàòÍ∞í Í≥ÑÏÇ∞

```python
from sklearn.metrics import mean_squared_error
```

```python
prediction = linearmodel.predict(X_test)
mse = mean_squared_error(y_test, prediction)
rmse = mse ** 0.5
print('mse : ',mse,'rmse : ', rmse)
```

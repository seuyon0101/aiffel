### 3.4.3 구현 정리 : 순전파 (forward propagation)

```python
def init_network():
    network = {}
    network['W1'] = np.array([[.1,.3,.5],[.2,.4,.6]])
    network['b1'] = np.array([.1,.2,.3])
    network['W2'] = np.array([[.1,.4],[.2,.5],[.3,.6]])
    network['b2'] = np.array([.1,.2])
    network['W3'] = np.array([[.1,.3],[.2,.4]])
    network['b3'] = np.array([.1,.2])
    
    return network
```

```python
def forward(network, x) :
    W1, W2, W3 = network['W1'], network['W2'], network['W3']
    b1, b2, b3 = network['b1'], network['b2'], network['b3']
    
    a1 = np.dot(x, W1) + b1
    Z1 = sigmoid(a1) # 위 정의
    a2 = np.dot(Z1, W2) + b2
    Z2 = sigmoid(a2)
    a3 = np.dot(Z2,W3) + b3
    y = identity_function(a3) #위 정의
    
    return y
```

```python
network = init_network()
x = [1.,.5]
y = forward(network, x)
print(y)
```

# 3.5 출력층 설계하기 

## 3.5.1 항등함수(identity function) & 소프트맥스 함수(softmax)

## 항등함수 :
## $$ \sigma () = y $$ 

## 소프트맥스 함수 :
## $$ y_k = \frac{exp(a_k)}{\sum_{i=1}^n exp(a_i)}$$

```python
def softmax(x) :
    exp = np.exp(x)
    sum_e = np.sum(exp)
    return exp / sum_e
```

## 3.5.2. 소프트맥스 구현시 주의점

```python
def csoftmax(x) :
    m = np.max(x)
    x = x - m
    exp = np.exp(x)
    sum_e = np.sum(exp)
    return exp / sum_e
```

```python
a = np.array([.3,2.9,4])
print(softmax(a))
print(csoftmax(a))
```

## 3.5.3. 소프트맥스 함수의 특징

```python
# softmax 총합은 1 or 100%
print(np.sum(softmax(a)))
```

## 3.6.2. 신경망의 추론 처리

```python
import pickle
def get_data() :
    (x_train, t_train),(x_test, t_test) = load_mnist(flatten=True, normalize=True, one_hot_label =False)
    return x_test, t_test

def init_network():
    with open('./sample_weight.pkl','rb') as f:
        network = pickle.load(f)
        return network
    
def predict(network, x) :
    W1, W2, W3 = network['W1'], network['W2'], network['W3']
    b1,b2,b3 = network['b1'], network['b2'], network['b3']
    
    a1 = np.dot(x,W1) + b1
    z1 = sigmoid(a1)
    a2 = np.dot(z1,W2) + b2
    z2 = sigmoid(a2)
    a3 = np.dot(z2,W3) + b3
    y = softmax(a3)
    
    return y
```

```python
x, t = get_data()
```

```python
network =init_network()
```

```python
accuracy_cnt= 0
```

```python
for i in range(len(x)):
    y = predict(network, x[i])
    p = np.argmax(y)
    if p == t[i]:
        accuracy_cnt += 1
        
print("accuracy : " + str(float(accuracy_cnt)/ len(x)))
```

## 3.6.3. 배치 처리

```python
x, _ = get_data()
network = init_network()

batch_size = 100
accuracy_cnt = 0

for i in range(0,len(x), batch_size):
    x_batch = x[i:i+batch_size]
    y_batch = predict(network, x_batch)
    p = np.argmax(y_batch, axis=1)
    y = predict(network, x_batch)
    p = np.argmax(y_batch, axis =1)
    accuracy_cnt += np.sum(p == t[i:i+batch_size])
        
print("accuracy : " + str(float(accuracy_cnt)/ len(x)))
```

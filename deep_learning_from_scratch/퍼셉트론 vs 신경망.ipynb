# 신경망 : 활성화 함수

## 1. 신경망 & 활성화 함수
입력 신호의 총합을 출력 신호로 변환하는 함수를 일반적으로 활성화 함수.

$$ a = b+ w_1x_1 + w_2x_2 + w_{n-1}x_{n-1} + w_{n}x_{n} $$

활성화 함수(step function; 계단식 함수) : $$y = h(a)$$

$$ h(x) = \{0(x \le 0) ; \\ 1(x > 0)\}$$

## 2. 활성화 함수

### 2.1 시그모이드 함수

$$ h(x) = \frac{1}{1+exp(-x)}$$

$$ exp(-x) = e^{-x} $$

```python
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
math.e ** 1
np.exp(1)
```

```python
def step_function(x):
    y = x >0
    return y.astype(np.int32)    
```

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
x = np.array([-1,1,2])
sigmoid(x)
```

```python
x = np.arange(-5,5,0.1)
y = sigmoid(x)
y2 = step_function(x)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
plt.plot(x,y)
plt.plot(x,y2)
plt.ylim(-0.1,1.1)
plt.title('sigmoid function')
plt.grid()
plt.show()
```

# 3. 비선형 함수

### 계단 함수와 시그모이드 함수의 공통점 = " 비선형 함수 "

> 선형 함수가 아니라 비선형 함수를 사용하는 이유는 **딥러닝 모델의 레이어 층을 깊게 가져갈 수 있기 때문이다.**

### Extra : 번외

```python
weight = [0.2,0.9,0.55,0.65,0.2,0.8,0.45,0.57]
target = [0,0,1,1,0,0,1,1]
```

```python
(0.5 * -34.4) + 2.14
```

```python
np.log(1 + np.exp(1)**0.3)
```

```python
def squiggle(x) :
    y1 = (x * -8.4) + 2.14
    y2 = (x * -2.52) + 1.29
    softplus1 = np.log(1 + np.exp(y1))
    softplus2 = np.log(1 + np.exp(y2))
    z1 = softplus1 * -1.3
    z2 = softplus2 * 2.28
    out = z1 + z2 -0.58
    return out
```

```python
x = np.arange(0,1,0.01)
y = squiggle(x)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
sns.scatterplot(x=weight,y= target, hue=target)
sns.lineplot(x=x,y=y)
plt.xlim(-0.1,1.1)
plt.grid()
plt.show()
```

```python
def relu(x) :
    return np.maximum(0,x)
```

### ReLu(Rectified Linear Unit ; 경사함수) 함수

```python
x = np.arange(-5,5,0.1)
y = relu(x)
plt.plot(x,y)
plt.title('ReLu activation function')
plt.grid()
plt.show()
```
